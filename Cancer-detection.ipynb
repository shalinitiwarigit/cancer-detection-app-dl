```python
import numpy as np
import pandas as pd
```

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout
from tensorflow.keras.optimizers import Adam
```

```python
train_path=r"C:\Users\Aman Tiwari\Desktop\Machine_Learning\train-cancer-detection"
test_path=r"C:\Users\Aman Tiwari\Desktop\Machine_Learning\test-cancer-detection"
```

```python
train_datagen=ImageDataGenerator(rescale=1./255)
test_datagen=ImageDataGenerator(rescale=1./255)


train_set=train_datagen.flow_from_directory(
    train_path,
    target_size=(150,150),
    class_mode='binary',
    batch_size=32 
)

test_set=test_datagen.flow_from_directory(
    test_path,
    target_size=(150,150),
    class_mode='binary',
    batch_size=32 
)
```

```python
#cnn structure
model=Sequential([
    Conv2D(32,(3,3),activation='relu',input_shape=(150,150,3)),
    MaxPooling2D(2,2),
    

    Conv2D(64,(3,3),activation='relu'),
    MaxPooling2D(2,2),
    

    Flatten(),
    Dense(128,activation='relu'),
    Dropout(0.5),
    Dense(1,activation='sigmoid')
    
])
```

```python
model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
model.fit(train_set,epochs=6,validation_data=test_set)
```

```python
from tensorflow.keras.preprocessing import image  # image is a module
import numpy as np

img_path = r"C:\Users\Aman Tiwari\Desktop\Machine_Learning\test-cancer-detection\test-no-cancer\0114.jpg"

def predictCancer(img_path):
    
    img = image.load_img(img_path, target_size=(150, 150)) 
    
    
    img_array = image.img_to_array(img) 
    
    
    img_array = np.expand_dims(img_array, axis=0)
    
    # 4. Normalize the image
    img_array /= 255.0

    # 5. Prediction
    predict = model.predict(img_array)  

    # 6. Interpret result
    if predict[0][0] < 0.5:
        print("Cancer")
    else:
        print("No Cancer")

# Call the function
predictCancer(img_path)
     
```
